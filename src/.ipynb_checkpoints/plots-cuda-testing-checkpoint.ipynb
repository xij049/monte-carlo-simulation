```python
# Read the data file
data_file = '../data/asian-options.txt'
with open(data_file) as f:
    paths = [[float(x) for x in l.split(',')] for l in f.readlines()]
```

```python
# Plot paths
def plot_paths(paths, dt, sample = 1.0):
    plt.figure()
    time = [i * dt for i in xrange(len(paths[0]))]
    for i in xrange(len(paths)):
        if rand() <= sample:
            plt.plot(time, paths[i])

# Plot final price distribution
def plot_prices(paths, sample = 1.0, bins = 50):
    endpoints = [paths[i][-1] for i in xrange(len(paths))]
    plt.figure()
    hist(endpoints, bins = 50)
```

```python
plot_paths(paths, dt = 0.01, sample = 0.01)
plot_prices(paths)
```
